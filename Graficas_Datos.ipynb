```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np; np.random.seed(0)
```

```python
df = pd.read_csv('heart_failure_clinical_records_dataset.csv')
df
```

```python
df['DEATH_EVENT'].value_counts().plot(kind='bar')
```

```python
sb.set_theme(style="whitegrid")
edadB = sb.boxplot(x=df['age'])
```

```python
df_fallecidos = df[df['DEATH_EVENT'] == 1]
df_Recuperados = df[df['DEATH_EVENT'] == 0]
```

```python
edadB_fallecidos = sb.boxplot(x=df_fallecidos['age'])
```

```python
edadB_Recuperados = sb.boxplot(x=df_Recuperados['age'])
```

```python
plt.plot(df_fallecidos['age'], df_fallecidos['ejection_fraction'], '*')
plt.plot(df_Recuperados['age'], df_Recuperados['ejection_fraction'], '*')
plt.title('Comparación de Fracción de eyección')
plt.xlabel("Edad")
plt.ylabel("Porcentaje (Fracción de Eyección)")
plt.legend(["Paciente fallecido", "Paciente recuperado"])
```

```python
plt.plot(df.age, df.serum_creatinine, '*')
plt.xlabel("Edad")
plt.ylabel("Anlisis de creatinina")
plt.show()
```

```python
df_val = df[['age','creatinine_phosphokinase', 'ejection_fraction', 'platelets', 'serum_creatinine', 'serum_sodium']]
df_valF = df_fallecidos[['age','creatinine_phosphokinase', 'ejection_fraction', 'platelets', 'serum_creatinine', 'serum_sodium']]
df_valR = df_Recuperados[['age','creatinine_phosphokinase', 'ejection_fraction', 'platelets', 'serum_creatinine', 'serum_sodium']]
```

```python
plt.figure(figsize=(11,11))
plt.title('Correlaciones entre atributos no booleanos', y=1.05, size=15)
sb.heatmap(df_val.corr(), linewidths=0.1, vmax=1.0, square=True, 
            linecolor='white', annot=True);
```

```python
plt.figure(figsize=(11,11))
plt.title('Correlaciones entre atributos no booleanos (Pacientes fallecidos)', y=1.05, size=15)
sb.heatmap(df_valF.corr(), linewidths=0.1, vmax=1.0, square=True, 
            linecolor='white', annot=True);
```

```python
plt.figure(figsize=(11,11))
plt.title('Correlaciones entre atributos no booleanos (Pacientes Recuperados)', y=1.05, size=15)
sb.heatmap(df_valR.corr(), linewidths=0.1, vmax=1.0, square=True, 
            linecolor='white', annot=True);
```

```python
plt.figure(figsize=(11,11))
plt.title('Correlaciones entre atributos', y=1.05, size=15)
sb.heatmap(df.corr(), linewidths=0.1, vmax=1.0, square=True, 
            linecolor='white', annot=True);
```
